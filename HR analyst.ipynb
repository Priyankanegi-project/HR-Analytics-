```python
import pandas as pd
from sqlalchemy import create_engine
```

```python
print("Running new code")
try:
    engine = create_engine("postgresql://postgres:1234@localhost:5432/hr_analyst")

    with engine.connect() as conn:
        print("Connection succeeded to PostgreSQL")

except Exception as e:
    print("Unable to connect")
    print("Error:", e)
```

```python
df.to_sql(
    "hr_analyst",
    con=engine,
    if_exists="replace",
    index=False
)

print("Table created")
```

```python
from sqlalchemy import text

with engine.connect() as conn:
    result = conn.execute(text("SELECT table_name FROM information_schema.tables WHERE table_schema='public';"))
    print(result.fetchall())
```

```python
from sqlalchemy import text

with engine.connect() as conn:
    result = conn.execute(text("SELECT * FROM hr_analyst LIMIT 5"))
    for row in result:
        print(row)
```
